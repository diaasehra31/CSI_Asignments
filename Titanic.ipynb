```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
titanic = sns.load_dataset("titanic")
```

```python
print(titanic.head())
```

```python
print(titanic.info())
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='survived', data=titanic)
plt.title('Survivors (0 = No, 1 = Yes)')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='sex', hue='survived', data=titanic)
plt.title('Survival by Gender')
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.histplot(titanic['age'].dropna(), kde=True, bins=30)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.heatmap(pd.crosstab(titanic['pclass'], titanic['survived']), annot=True, fmt='d', cmap='YlGnBu')
plt.title('Survival by Class')
plt.show()

```

```python
sns.pairplot(titanic[['survived', 'age', 'fare', 'pclass']], hue='survived')
plt.suptitle('Pairplot of Titanic Features', y=1.02)
plt.show()
```
